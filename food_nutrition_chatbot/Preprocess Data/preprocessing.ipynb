```python
import pandas as pd
```

```python
df2 = pd.read_csv("NutritionalFacts_Fruit_Vegetables_Seafood.csv",encoding='latin-1')
df2
```

```python
df1 = df2.drop(index=0)
```

```python
import numpy as np
df = df1.replace(np.nan, 0)
df
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
final_df = df.rename(columns={'Food and Serving': 'Food_item',
                        'CaloriesÊfrom Fat': 'Fat_Calories',
                        'Total Fat': 'Total_Fat_g',
                        'Total Fat.1' : 'Total_Fat_DV',
                        'Sodium':'Sodium_g',
                        'Sodium.1':'Sodium_DV',
                        'Potassium':'Potassium_g',
                        'Potassium.1':'Potassium_DV',
                        'Total Carbo-hydrate':'Total_Carbohydrate_g',
                        'Total Carbo-hydrate.1':'Total_Carbohydrate_DV',
                        'Dietary Fiber':'Dietary_Fiber_g',
                        'Dietary Fiber.1':'Dietary_Fiber_DV',
                        'Vitamin A':'Vitamin_A',
                        'Vitamin C':'Vitamin_C',
                        'ÊÊIronÊÊ':'Iron',
                        'Saturated Fat':'Saturated_Fat_g',
                        'Saturated Fat.1':'Saturated_Fat_DV',
                        'Chole-sterol':'Cholesterol_g',
                        'Chole-sterol.1':'Cholesterol_DV',
                        'Food Type':'Food_Type'    
                       })
final_df
```

```python
# df["Food_Type"] = df['Food_Type'].map({'Fake':0, 'Real':1})

# Vegetables, Serving Size (gram weight/Êounce weight)
# Fruits ServingÊSize (gramÊweight/ounceÊweight)
# Seafood, Serving Size (84 g/3 oz)
```

```python
# Dictionary for mapping old values to new values
mapping = {
    'Vegetables, Serving Size (gram weight/Êounce weight)': 'Vegetables',
    'Fruits ServingÊSize (gramÊweight/ounceÊweight)': 'Fruits',
    'Seafood, Serving Size (84 g/3 oz)': 'Seafood'
}

# Replace values using the dictionary
final_df['Food_Type'] = final_df['Food_Type'].replace(mapping)

print(final_df)
```

```python
final_df
```

```python
final_df.to_csv('food_nutrition.csv', index=False, encoding = 'utf-8')
```
